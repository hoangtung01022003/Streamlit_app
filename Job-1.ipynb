## 1.      Số lượng người dùng và mức lương từng ngôn ngữ

```python
# Tính toán số lượng người dùng cho mỗi ngôn ngữ
language_counts = df['language'].value_counts()

# Tính mức lương trung bình cho mỗi ngôn ngữ
average_salary = df.groupby('language')['salary'].mean()

# Tạo biểu đồ
fig, ax1 = plt.subplots()

# Biểu đồ cột cho số lượng người dùng
language_counts.plot(kind='bar', color='blue', ax=ax1, width=0.4, position=1)
ax1.set_ylabel('Số lượng người dùng', color='blue')

# Tạo trục thứ hai cho mức lương trung bình
ax2 = ax1.twinx()
average_salary.plot(kind='bar', color='red', ax=ax2, width=0.4, position=0)
ax2.set_ylabel('Mức lương trung bình ($)', color='red')

# Hiển thị biểu đồ
plt.title('Số lượng người dùng và mức lương trung bình theo ngôn ngữ')
plt.show()
```

##2.      Sự phổ biến của từng nhóm Job – Sử dụng Pie

```python
curent_role_value = df['curent_role'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(curent_role_value, labels=curent_role_value.index, autopct='%1.1f%%', startangle=140)
plt.title('Phổ biến của các nhóm công việc')
plt.axis('equal')  # Biểu đồ tròn sẽ có hình dạng hình tròn
plt.show()
```

## 3.      Mức lương theo từng nhóm Job



```python

# Phân loại dữ liệu theo từng current_role và tính mức lương trung bình
average_salary_by_role = df.groupby('curent_role')['salary'].mean().reset_index()

# Tạo biểu đồ cột
plt.figure(figsize=(10, 6))
bar = plt.bar(average_salary_by_role['curent_role'], average_salary_by_role['salary'], color='skyblue')
for bar in bar:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom', ha='center')
plt.xlabel('Nhóm công việc')
plt.ylabel('Mức lương')
plt.title('Mức lương theo từng nhóm công việc')
plt.xticks(rotation=45, ha='right')  # Xoay nhãn trục x để dễ đọc
plt.tight_layout()
plt.show()
```

## 4.      Mức độ Happy với từng Job – Sử dụng Scatter

```python
average_happiness_by_job = df.groupby('curent_role')['happy_salary'].mean().reset_index()

# Tạo biểu đồ scatter
plt.figure(figsize=(10, 6))
plt.scatter(average_happiness_by_job['curent_role'], average_happiness_by_job['happy_salary'], color='skyblue', s=100)

# Thêm nhãn cho trục x và y
plt.xlabel('Nhóm công việc')
plt.ylabel('Mức độ Happy')
plt.title('Mức độ Happy theo từng nhóm công việc')

# Đặt các nhãn trục x theo góc 90 độ để dễ đọc hơn
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
print(df.columns)
```

## 5.      Mức lương trung bình theo từng nhóm job của từng ngôn ngữ

```python
average_sal_by_job_and_language = df.groupby(['curent_role', 'language'])['salary'].mean().reset_index()

# Tạo biểu đồ cột
plt.figure(figsize=(12, 8))

# Duyệt qua từng ngôn ngữ và vẽ cột cho mỗi ngôn ngữ
for language in average_sal_by_job_and_language['language'].unique():
    data = average_sal_by_job_and_language[average_sal_by_job_and_language['language'] == language]
    plt.bar(data['curent_role'], data['salary'], label=language, alpha=0.7)

# Thêm nhãn cho trục x và y, và tiêu đề
plt.xlabel('Nhóm công việc')
plt.ylabel('Mức lương')
plt.title('Mức lương trung bình theo từng nhóm công việc của từng ngôn ngữ')

# Đặt các nhãn trục x theo góc 90 độ để dễ đọc hơn
plt.xticks(rotation=0)

# Hiển thị chú thích
plt.legend()

plt.tight_layout()
plt.show()
```
