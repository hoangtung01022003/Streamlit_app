#### Xử lý dữ liệu

#### Thay rỗng bằng NA

```python
df = df.replace(r'^s*$',np.nan)
```

```python
#Kiểm tra DL trong cột 'current_role'
df['curent_role'].unique()
#df['curent_role'].nunique()
```

#### Cột 'curent_role' chứa rất nhiều giá trị khác hau. Gom các Other... thành 1 loại other

```python
df['curent_role']=df['curent_role'].str.replace('Other.*','Other',regex=True)
```

```python
df['curent_role'].unique()
```

```python
df.columns
```

```python
#Kiểm tra cột salary
```

```python
df['salary'].unique()
```

#### Tách cột salary thành 2 cột khác nhau

```python
df[['sal_min','sal_max']]=df['salary'].str.split('-',n=2,expand=True)
```

```python
df.head(1)
```

```python
df['sal_min']=df['sal_min'].str.extract('(\d+)')
df['sal_max']=df['sal_max'].str.extract('(\d+)')
```

```python
df.head(1)
```

```python
df['sal_max'].unique()
```

```python
#Trường hợp sal_max = NA vi 225k+
df.loc[df['sal_max'].isna(),'sal_max']=df['sal_min']
```

```python
df['sal_max'].unique()
```

```python
df[['sal_max','sal_min']]=df[['sal_max','sal_min']].astype(float)
```

```python
df.info()
```

```python
df['salary']=(df['sal_min']+df['sal_max'])/2
#sau đó thì drop 2 cột min max
```

```python
df['language']=df['language'].str.replace('Other.*','Other',regex=True)
```

```python
df.loc[df['age']>100,'age']
df['age'].mean()
```
